```python
import pandas as pd
```

```python
df = pd.read_csv('data/titanic.csv')
df
```

```python
df.set_index('PassengerId', inplace = True)
```

```python
df=df[['Survived', 'Pclass', 'Sex', 'Age']]
df
```

```python
df.info()
```

```python
df.describe()
```

```python
survived_1 = df['Survived'] == 1
sex_1 = df['Sex'] == 'male'
df_1 = df[survived_1 & sex_1]

survived_2 = df['Survived'] == 1
sex_2 = df['Sex'] == 'female'
df_2 = df[survived_2 & sex_2]

survived_3 = df['Survived'] == 0
sex_3 = df['Sex'] == 'male'
df_3 = df[survived_3 & sex_3]

survived_4 = df['Survived'] == 0
sex_4 = df['Sex'] == 'female'
df_4 = df[survived_4 & sex_4]
```

```python
df_1
```

```python
df_1.info()
```

```python
df_1['Age'].mean()
```

```python
df_1['Age'] = df_1['Age'].fillna(df_1['Age'].mean())
df_2['Age'] = df_2['Age'].fillna(df_2['Age'].mean())
df_3['Age'] = df_3['Age'].fillna(df_3['Age'].mean())
df_4['Age'] = df_4['Age'].fillna(df_4['Age'].mean())
```

```python
df_res = df_1.append(df_2)
df_res = df_res.append(df_3)
df_res = df_res.append(df_4)
```
