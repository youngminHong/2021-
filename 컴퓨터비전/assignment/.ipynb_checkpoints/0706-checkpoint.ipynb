```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
point = np.array([[-2.9, 35.4],
                 [-2.1, 19.7],
                 [-0.9, 5.7],
                 [0.1, 1.2],
                  [1.1, 2.1],
                 [1.9, 8.7],
                 [3.1, 25.7],
                 [4.0, 41.5]])
point.shape
```

```python
x = point[:,0]
y = point[:,1]

plt.scatter(x,y, c = 'black', s= 100)
plt.show()
```

```python
a, b, c = np.polyfit(x, y, 2)
print(a, b, c)
```

```python
fit_x = np.arange(-3,4,0.1)
fit_y = a*fit_x**2 + b*fit_x + c
    
plt.scatter(x,y, c = 'black', s= 100)
plt.plot(fit_x, fit_y)
plt.title(f"{a}x^2 + {b}x + {c}")
plt.show()
```
